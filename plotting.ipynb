```julia
using Plots, JLD2, Measures
```

```julia
res = load("saved_networks/network_robustness_er_A.jld2")
```

```julia
y_deg = res["y_deg"]
y_rand = res["y_rand"]
y_between = res["y_between"]
y_closeness = res["y_closeness"]
f = res["f"]
```

```julia
pl1 = plot(f, y_rand[1, :], title = "k = 0.5", label = "", ylabel = "size of the giant component", legend = :topright, 
markershape = :circle)
plot!(pl1, f, y_deg[1, :], label = "", markershape = :circle)

pl2 = plot(f, y_rand[2, :], title = "k = 1", label = "random", legend = :topright, 
markershape = :circle)
plot!(pl2, f, y_deg[2, :], label = "degree", markershape = :circle)

pl3 = plot(f, y_rand[3, :], title = "k = 2", label = "", ylabel = "size of the giant component", xlabel = "fraction of removed nodes", legend = :topright, 
markershape = :circle)
plot!(pl3, f, y_deg[3, :], label = "", markershape = :circle)

pl4 = plot(f, y_rand[4, :], title = "k = 4", label = "", xlabel = "fraction of removed nodes", legend = :topright, 
markershape = :circle)
plot!(pl4, f, y_deg[4, :], label = "", markershape = :circle)

plot(pl1, pl2, pl3, pl4, layout = (2, 2), size = (1300, 900), left_margin = 4mm)
```

Fig. 1. Size of the giant component (normalized to the size of the giant component before removal of nodes) with respect to the fraction of removed nodes for two different attack strategies: random and degree-based. Results for Erdos-Renyi $G(n,p)$ model with $n = 10^5$ and four different $\langle k \rangle$. For each fraction $L = 10^2$ realisations were computued.
