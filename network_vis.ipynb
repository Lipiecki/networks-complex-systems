```julia
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```julia
G = nx.read_weighted_edgelist("real_networks/email-Eu-core-temporal.txt", create_using=nx.Graph(), nodetype=int)
```

```julia
edge_w = nx.get_edge_attributes(G, 'weight')
edge_colors = [edge_w[e] for e in G.edges()]
```

```julia
G = nx.read_edgelist("real_networks/CA-GrQc.txt", create_using=nx.Graph(), nodetype=int)
deg_centrality = nx.degree_centrality(G)
maxdeg = np.log(max(deg_centrality.values()))
node_colors = [np.log(deg_centrality[v]) for v in sorted(G)]
```

```julia
nx.draw(G, pos=nx.spring_layout(G, dim = 2, seed = 520), node_size=5, with_labels=False, node_color = node_colors, cmap="plasma", edge_color = "gray")
```

```julia
pos = nx.spring_layout(G, dim=3, seed=712)

node_xyz = np.array([pos[v] for v in sorted(G)])
edge_xyz = np.array([(pos[u], pos[v]) for u, v in G.edges()])

fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")

ax.scatter(*node_xyz.T, s=25, ec="w", c = node_colors, cmap = "plasma", alpha = 0.9, edgecolors="face")

for vizedge in edge_xyz:
    ax.plot(*vizedge.T, color="tab:gray", linewidth=0.25)


def _format_axes(ax):
    """Visualization options for the 3D axes."""
    # Turn gridlines off
    ax.grid(False)
    # Suppress tick labels
    for dim in (ax.xaxis, ax.yaxis, ax.zaxis):
        dim.set_ticks([])
    # Set axes labels


_format_axes(ax)
fig.tight_layout()

plt.rcParams['figure.figsize'] = [15, 15]
plt.show()
```
